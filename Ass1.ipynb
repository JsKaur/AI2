```python
diabetes['SkinThickness'].replace(0, np.nan, inplace=True)
diabetes['Insulin'].replace(0, np.nan, inplace=True)
diabetes['BMI'].replace(0, np.nan, inplace=True)
diabetes['BloodPressure'].replace(0, np.nan, inplace=True)
```

```python
median_value = diabetes['Insulin'].median()
rounded_median = round(median_value, 1)
diabetes['Insulin'].fillna(rounded_median, inplace=True)
```

```python
median_value = diabetes['SkinThickness'].median()
rounded_median = round(median_value, 1)
diabetes['SkinThickness'].fillna(rounded_median, inplace=True)
```

```python
median_value = diabetes['BMI'].median()
rounded_median = round(median_value, 1)
diabetes['BMI'].fillna(rounded_median, inplace=True)
```

```python
median_value = diabetes['BloodPressure'].median()
rounded_median = round(median_value, 1)
diabetes['BloodPressure'].fillna(rounded_median, inplace=True)
```

```python
X, y = diabetes.drop('Outcome', axis=1), diabetes['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=87)
```

```python
from sklearn.linear_model import LogisticRegression

logmodel = LogisticRegression(max_iter=780)
logmodel.fit(X_train,y_train)
```

```python
logmodel.score(X_test, y_test)
```
